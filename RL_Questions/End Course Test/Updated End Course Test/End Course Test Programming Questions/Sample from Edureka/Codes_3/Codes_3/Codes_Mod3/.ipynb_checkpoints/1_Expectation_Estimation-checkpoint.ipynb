# Expectation

The expected value (or mean) of X, where X is a discrete random variable, is a weighted average of the possible values that X can take, each value being weighted according to the probability of that event occurring. The expected value of X is usually written as E(X).


A random variable is some event whose value is determined by randomness. Expected value gives us an idea on what to expect when we repeat the event for multiple times.


```python
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline
```

Lets consider a fair dice that has equal chance of turning up one of its six faces. Below we compute the expected value for it.

```python
expected_value = 1/6.0 *(1+2+3+4+5+6)
print(expected_value)
```

```python
def estimate_expected_value(num_throws):

    
    sum_of_dice_values = 0
    running_average=[]
    for i in range(num_throws):
        dice_value = np.random.randint(1,7)
        sum_of_dice_values += dice_value
        running_average.append(sum_of_dice_values/(i+1.0))
    sum_of_dice_values/ num_throws
    running_average = np.asarray(running_average)
    running_average
    average = sum_of_dice_values /num_throws
    print(average)
    x= [i for i in range(num_throws)]
    y= [average for i in range(num_throws)]
    plot.plot(running_average)
    plot.plot(x,y,'k-')
    plot.show()
```

Lets repeat the experiment with 10 trials

```python
estimate_expected_value(10)
```

```python
estimate_expected_value(100)
```

```python
estimate_expected_value(1000)
```

```python
estimate_expected_value(10000)
```
