```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%pylab inline
import random
```

```python
gamma = 0.9 # discounting rate
rsize = -1
Grid = 11
TState = [[0,10], [6,4]]
BState =[[0,5],[1,5],[3,5],[4,5],[5,5],[6,5],[7,5],[8,5],[10,5],[5,0],[5,2],[5,3],[5,4],[6,6],[6,7],[6,9],[6,10]]
NState =[[0,9],[1,8],[1,9],[2,7],[2,8],[3,9],[4,7],[5,10]]
actions = [[1, 0], [-1, 0], [0, 1], [0, -1]]
It = 500
```

```python
def ARFunction(StartP, action):
    
    if StartP in TState:
        return StartP, 0
    if StartP in BState:
        return StartP, 0
    if StartP in NState:
        return StartP, 0
    
    r = rsize
    finalP = np.array(StartP) + np.array(action)
    if -1 in finalP or 11 in finalP: 
        finalP = StartP
        
    return finalP, r
```

```python
vMap = np.zeros((Grid, Grid))
states = [[i, j] for i in range(Grid) for j in range(Grid)]
```

```python
# values of the value function at step 0
vMap
```

```python
delta = []
for it in range(It):
    cpyVMap = np.copy(vMap)
    deltaS = []
    for s in states:
        wR = 0
        for a in actions:
            finalP, r = ARFunction(s, a)
            wR += (1/len(actions))*(r+(gamma*vMap[finalP[0], finalP[1]]))
        deltaS.append(np.abs(cpyVMap[s[0], s[1]]-wR))
        cpyVMap[s[0], s[1]] = wR
    delta.append(deltaS)
    vMap = cpyVMap
    if it in [0,1,2,9, 99, It-1]:
        print("Iteration {}".format(it+1))
        print(vMap)
        print("")
```
